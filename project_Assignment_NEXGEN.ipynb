```python
#2. **Genre Insights**

#`Which movie genres are the most frequently rated?`

#`Compare the average ratings across different genres. Are certain genres consistently rated higher or lower?`

# Step 1: Read the movies data
with open(r'C:\Users\deoka\OneDrive\Desktop\ml-1m\ml-1m\movies.dat') as movies:
    movie_genres = {}
    for line in movies:
        line = line.strip()
        columns = line.split('::')
        movie_id = int(columns[0])
        genres = columns[2].split('|')
        movie_genres[movie_id] = genres

# Step 2: Read the ratings data
with open(r'C:\Users\deoka\OneDrive\Desktop\ml-1m\ml-1m\ratings.dat') as ratings:
    movie_rating = {}
    for line in ratings:
        line = line.strip()
        columns = list(map(int, line.split('::')))
        movie_id = columns[1]
        rating = columns[2]
        if movie_id in movie_rating:
            movie_rating[movie_id].append(rating)
        else:
            movie_rating[movie_id] = [rating]

# Step 3: Map movie IDs to their genres and calculate frequencies and ratings
genre_frequency = {}
genre_rating = {}
for movie_id, genres in movie_genres.items():
    if movie_id in movie_rating:
        for genre in genres:
            if genre in genre_frequency:
                genre_frequency[genre] += len(movie_rating[movie_id])
                genre_rating[genre].extend(movie_rating[movie_id])
            else:
                genre_frequency[genre] = len(movie_rating[movie_id])
                genre_rating[genre] = movie_rating[movie_id][:]

# Step 4: Calculate average ratings for each genre
genre_average_ratings = {}
for genre, rating in genre_rating.items():
    genre_average_ratings[genre] = sum(rating) / len(rating)

# Step 5: Find the genre with the highest frequency
most_frequently_rated_genre = max(genre_frequency, key=genre_frequency.get)
print(f'The most frequently rated genre is: {most_frequently_rated_genre}, with {genre_frequency[most_frequently_rated_genre]} ratings')

# Step 6: Find the genre with the highest and lowest average ratings
highest_rated_genre = max(genre_average_ratings, key=genre_average_ratings.get)
lowest_rated_genre = min(genre_average_ratings, key=genre_average_ratings.get)

print(f'The genre with the highest average rating is: {highest_rated_genre}, with an average rating of {genre_average_ratings[highest_rated_genre]:.2f}')
print(f'The genre with the lowest average rating is: {lowest_rated_genre}, with an average rating of {genre_average_ratings[lowest_rated_genre]:.2f}')

# Step 7: Compare genre ratings
for genre, avg_rating in genre_average_ratings.items():
    print(f'Genre: {genre}, Average Rating: {avg_rating:.2f}')



      
```

```python
#6. **Exploring Long Tail**

#`Investigate the "long tail" of the dataset: How many movies receive very few ratings?`

#`What are the characteristics of these less-rated movies compared to popular ones?`
# Step 1: Read movie data
with open(r'C:\Users\deoka\OneDrive\Desktop\ml-1m\ml-1m\movies.dat') as movies:
    movie_data = {}
    for line in movies:
        line = line.strip()
        columns = line.split('::')
        movie_id = int(columns[0])
        title = columns[1]
        genres = columns[2].split('|')
        release_year = int(title.strip()[-5:-1])
        movie_data[movie_id] = {'title': title, 'release_year': release_year, 'genres': genres}

# Step 2: Read ratings data
with open(r'C:\Users\deoka\OneDrive\Desktop\ml-1m\ml-1m\ratings.dat') as ratings:
    movie_ratings_count = {}
    for line in ratings:
        line = line.strip()
        columns = list(map(int, line.split('::')))
        movie_id = columns[1]
        if movie_id in movie_ratings_count:
            movie_ratings_count[movie_id] += 1
        else:
            movie_ratings_count[movie_id] = 1

# Step 3: Identify movies with few ratings
few_ratings_threshold = 10  # Example threshold for few ratings
few_ratings_movies = {movie_id: count for movie_id, count in movie_ratings_count.items() if count <= few_ratings_threshold}
popular_movies = {movie_id: count for movie_id, count in movie_ratings_count.items() if count > few_ratings_threshold}

# Step 4: Compare characteristics
def analyze_movie_characteristics(movie_ids, movie_data):
    release_years = [movie_data[movie_id]['release_year'] for movie_id in movie_ids]
    genres = [genre for movie_id in movie_ids for genre in movie_data[movie_id]['genres']]
    return release_years, genres

few_ratings_years, few_ratings_genres = analyze_movie_characteristics(few_ratings_movies.keys(), movie_data)
popular_years, popular_genres = analyze_movie_characteristics(popular_movies.keys(), movie_data)

from collections import Counter
few_ratings_years_counter = Counter(few_ratings_years)
few_ratings_genres_counter = Counter(few_ratings_genres)
popular_years_counter = Counter(popular_years)
popular_genres_counter = Counter(popular_genres)

# Print results
print("Less-Rated Movies Characteristics:")
print("Release Years:", few_ratings_years_counter)
print("Genres:", few_ratings_genres_counter)

print("\nPopular Movies Characteristics:")
print("Release Years:", popular_years_counter)
print("Genres:", popular_genres_counter)

```
